```python
# linear Regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([[1],[2],[3],[4],[5]]) # input hours
y = np.array([1,2,3,3,5]) # marks
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
print("slope(m)",model.coef_)
print("intercept",model.intercept_)
```

```python
plt.scatter(x,y,color='red',label="Actual data")
plt.plot(x,y_pred,color = "blue" , label="BEST fit line")
plt.xlabel("HOURS STUDIED")
plt.ylabel("MARKS OBTAINED")
plt.legend()
plt.show()
```

```python
print("predect for 6 hours study", model.predict([[12]]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge,Lasso
```

```python
# lasso ridge regression
# data banana
x = np.array([[1],[2],[3],[4],[5],[6]]) # input hours
y = np.array([1,2,1.3,3.75,2.5,5]) # marks
```

```python
# teen modle banye
lin_model = LinearRegression()
ridge_model = Ridge(alpha = 1.0)
lasso_model = Lasso(alpha = 0.1)

```

```python
# step 3 : trainning karna(fit karana)
lin_model.fit(x,y)
ridge_model.fit(x,y)
lasso_model.fit(x,y)
```

```python
# step 4 predect
y_lin = lin_model.predict(x)
y_ridge = ridge_model.predict(x)
y_lasso = lasso_model.predict(x)
```

```python
plt.scatter(x,y,color ="red",label ="actual data")
plt.plot(x,y,label="linear",color="blue")
plt.plot(x,y_ridge,label="ridge",color="green")
plt.plot(x,y_lasso,label="lasso",color = "purple") # real data poits
plt.legend()
plt.show()
```

```python
print("predect for 12 hours study", ridge_model.predict([[6]]))
print("predect for 12 hours study", lasso_model.predict([[6]]))
print("predect for 12 hours study", lin_model.predict([[6]]))
```
